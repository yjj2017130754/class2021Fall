```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
def phasor_sine(amp, sr, dur, freq, phase):
  t = np.arange(1, sr * dur+1)/sr
  theta = t * 2*np.pi * freq + phase
  s = amp * np.sin(theta)
  return s, t, theta
```

```python
amp = 1; sr = 10000; phase = 0

C  = 523.25
Db = 554.37
D  = 587.33
Eb = 622.25
E  = 659.25	  
F  = 698.46
Gb = 739.99 
G  = 783.99
Ab = 830.61
A  = 880.00
Bb = 932.33
B  = 987.77
```

```python
# 엘리제를 위하여

notes = [E*2, Db*2, E*2, Db*2, E*2, B, D*2, C*2, A, 0, 
         C, E, A, B, 0, E, Gb, B, C*2, 0, 
         E, E*2, Db*2, E*2, Db*2, E*2, B, D*2, C*2, A, 0,
         C, E, A, B, 0, D, C*2, B, A, 0,
         B, C*2, D*2, E*2, G, F*2, E*2, D*2, F, E*2, D*2, C*2, E, D*2, C*2, B, 0]

durs = [0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.5, 0.25, 
        0.25, 0.25, 0.25, 0.5, 0.25, 0.25, 0.25, 0.25, 0.5, 0.25,
        0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.5, 0.25,
        0.25, 0.25, 0.25, 0.5, 0.25, 0.25, 0.25, 0.25, 0.5, 0.25,
        0.25, 0.25, 0.25, 0.75, 0.25, 0.25, 0.25, 0.75, 0.25, 0.25, 0.25, 0.75, 0.25, 0.25, 0.25, 0.5, 0.25]
```

```python
s_all = np.empty(0)

for note, dur in zip(notes, durs):
  s, t, th= phasor_sine(amp, sr, dur, note, phase)
  s_all = np.concatenate((s_all, s))
```
